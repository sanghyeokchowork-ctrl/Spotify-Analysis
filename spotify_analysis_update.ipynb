```python
#Artist Info Collection
feature_list = []
print(f"\n Step 4: Artist Info Collection")

for artist_name, artist_id in artist_map.items():
    try:
        print(f"Processing: {artist_name} (ID: {artist_id})")

        # API Call 1: Get Profile
        artist_profile = sp.artist(artist_id)
        # API Call 2: Get Albums
        artist_albums = sp.artist_albums(artist_id=artist_id, limit=50)

        # Get the data
        data_to_save = {
            'name': artist_name,
            'id': artist_id,
            'popularity': artist_profile['popularity'],
            'genres': artist_profile['genres'],
            'followers': artist_profile['followers']['total'],
            'album_count': len(artist_albums['items'])
        }

        # APPEND this dictionary to your main list
        feature_list.append(data_to_save)

        # This is the success message we want to see!
        print(f"  > Success! Popularity: {artist_profile['popularity']}, Followers: {artist_profile['followers']['total']}")
        time.sleep(0.3) # API rate limiting

    except Exception as e:
        print(f"  !! Error processing {artist_name}: {e}")

#Show The Result
print("\n--- Artist Info Collection Completed! ---")
print(f"Total artists collected: {len(feature_list)}")

if not feature_list:
    print("--- No artist data was collected. ---")
else:
    print("\n---  Displaying All Collected Artist Data ️ ---")

    # Loop through each item in the feature_list
    for i, artist_data in enumerate(feature_list):

        # Get the artist name from the collected data for a nice header
        artist_name = artist_data.get('name', f'Artist {i+1}')

        print(f"\n--- {i+1}. {artist_name} ---")

        # Use pprint to print the dictionary cleanly
        pprint.pprint(artist_data)

        print("-" * 30) # Add a separator
```

```python
import pprint

genre_map = {}

# Check if feature_list exists
if 'feature_list' not in locals() or not feature_list:
    print(" 'feature_list' not found or is empty.")
    print("Please run Cell 4 (Artist Info Collection) successfully first.")
else:
    print(f"Processing {len(feature_list)} artists...")

    # Loop through each artist we collected in Cell 4
    for artist_data in feature_list:
        artist_name = artist_data['name']
        genres = artist_data['genres']

        # Loop through the list of genres for that artist
        for genre in genres:
            # If this is the first time we see this genre, create an empty list
            if genre not in genre_map:
                genre_map[genre] = []

            # Add the artist's name to this genre's list
            genre_map[genre].append(artist_name)

    print("\n--- ️ Genre Classification Complete! ️ ---")
    print(f"Found {len(genre_map)} unique genres across {len(feature_list)} artists.")

    print("\n(Genre -> Artist List):")
    pprint.pprint(genre_map)
```
